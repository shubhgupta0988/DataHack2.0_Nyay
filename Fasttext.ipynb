```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/data.csv")
```

```python
data.head()
```

```python
texts = []

for item in data['sentence']:
  texts.append(item)
```

```python
import fasttext

# Specify your training data and model parameters

input_file = 'input.txt'

# Convert the DataFrame column to a list and save it to a text file with each row on a new line
column_data = data['sentence'].tolist()
with open(input_file, 'w') as file:  # Use a different variable name for the file object
    file.write('\n'.join(column_data))

output_model = 'model.bin'
model = fasttext.train_unsupervised(input=input_file, model='skipgram', lr=0.1, dim=100)

# Save the trained model
model.save_model(output_model)
```

```python
model.save_model('fasttext.bin')
```

```python
from annoy import AnnoyIndex

# Convert text data into embeddings
text_data = texts
vector_data = [model.get_sentence_vector(text) for text in text_data]

# Create and build an Annoy index
def build_annoy_index(vector_data, index_filename):
    t = AnnoyIndex(len(vector_data[0]), metric='euclidean')
    for i, vector in enumerate(vector_data):
        t.add_item(i, vector)
    t.build(1000)  # Adjust the number of trees as needed
    t.save(index_filename)

index_filename = 'annoy_index.ann'
build_annoy_index(vector_data, index_filename)

# Query the Annoy index with a new text query
query_text = "hi dear sir @ madam i am work enterprises shop casher my owner my big bother my owner allcosmer money take and leave in city all cosmer tarcher and attack me and my family what i do sir"
query_vector = model.get_sentence_vector(query_text)

t = AnnoyIndex(len(query_vector), metric='euclidean')
t.load(index_filename)
num_neighbors = 5
nearest_neighbors = t.get_nns_by_vector(query_vector, num_neighbors)

# Print the nearest neighbors
for neighbor_index in nearest_neighbors:
    print(f"Vector {neighbor_index}: {text_data[neighbor_index]}")
```
